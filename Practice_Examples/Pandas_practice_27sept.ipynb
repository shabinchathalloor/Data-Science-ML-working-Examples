```python
df1 = pd.DataFrame({'A':['A0','A1','A2','A3'],
                   'B':['B0','B1','B2','B3'],
                   'C':['C0','C1','C2','C3'],
                   'D':['D0','D1','D2','D3']},
                  index = [0,1,2,3])
df2 = pd.DataFrame({'A':['A4','A5','A6','A7'],
                   'B':['B4','B5','B6','B7'],
                   'C':['C4','C5','C6','C7'],
                   'D':['D4','D5','D6','D7']},
                  index = [4,5,6,7])
df3 = pd.DataFrame({'C':['A8','A9','A10','A11'],
                   'B':['B8','B9','B10','B11'],
                   'C':['C8','C9','C10','C11'],
                   'D':['D8','D9','D10','D11']},
                  index = [8,9,10,11])
frames = [df1,df2,df3]
result = pd.concat(frames,sort=False)
print(result)
```

```python
s1 = pd.Series(['a','b','c'])
s2 = pd.Series(['c','d','c','d','e'])
con1=pd.concat([s1,s2]) #combnine two series
print("Concatination of S1&S2")
print(con1)
con2=pd.concat([s1,s2],ignore_index=True)
#clear the existing index and reset it in the result by setting the ignore_index option to True.
print();print("Concatination of S1&S2 with ignoring index")
print(con2)
#add a hierarchical index at the outermost level of the data with key option
con3=pd.concat([s1,s2],keys=['s1', 's2'],)
print();print(con3)
#label the index key you crate with the names option
pd.concat([s1,s2],keys=['s1','s2'],
         names=['series name','Row ID'])
```

```python
#combine to DataFrame objects with identical columns
Df1 = pd.DataFrame([['a',1],['b',2]],
                  columns=['letter','number'])
Df2 = pd.DataFrame([['c',3],['d',4]],
                  columns=['letter','number'])
pd.concat([Df1,Df2])
```

```python
#columns outside the intersection will be filled with NaN values
Df3 = pd.DataFrame([['c',3,'cat'],['d',4,'dog']],
                  columns=['letter','number','animal'])
print(Df3)
print()
result=pd.concat([Df1,Df3], sort=False)
print(result);print()
Result=pd.concat([Df1,Df3],join="inner")
print(Result)
```
