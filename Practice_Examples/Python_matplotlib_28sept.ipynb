```python
x=np.arange(0,25)
y=x*2
fig,axes = plt.subplots(nrows=1, ncols=2, figsize=(12,4))

axes[0].plot(x,y,color="blue", lw=1)
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')

axes[1].plot(x,z,color="red", lw=3, ls='-')
axes[1].set_xlabel('x')
axes[1].set_ylabel('z')
```

```python
hours=[4,8,2]
activities=['sleep','work','eat']
colors=['r','g','b']

plt.pie(hours,
       labels=activities,
       colors=colors,
       startangle=90,
       autopct='%.2f%%')
plt.show()
```
